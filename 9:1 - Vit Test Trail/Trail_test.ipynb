```python
from transformers import ViTForImageClassification, ViTFeatureExtractor

model = ViTForImageClassification.from_pretrained("shng2025/trash_vit_trail")
feature_extractor = ViTFeatureExtractor.from_pretrained("shng2025/trash_vit_trail")
```

```python
from PIL import Image
import requests
import torch
from transformers import ViTFeatureExtractor, ViTForImageClassification
import matplotlib.pyplot as plt
import torch.nn.functional as F
import time

def classify_image(image_url, output_type="both"):
    """
    Classify an image using a Vision Transformer and display softmax probabilities, raw logits, or both.

    Args:
        image_url (str): URL of the image to classify.
        output_type (str): "softmax", "none", or "both" to specify how to handle the model output.

    Returns:
        None: Displays the image and prints the predicted class label with the selected output type.
    """
    # Load and prepare the image
    image = Image.open(requests.get(image_url, stream=True).raw)
    if image.mode != 'RGB':
        image = image.convert('RGB')  # Convert grayscale to RGB

    print(f"Image resolution: {image.size}")  # Print image resolution

    inputs = feature_extractor(images=image, return_tensors="pt")

    # Inference
    model.eval()
    start_time = time.time()  # Start time
    with torch.no_grad():
        outputs = model(**inputs)
        logits = outputs.logits
    end_time = time.time()  # End time

    # Calculate and print inference time
    inference_time = end_time - start_time
    print(f"Inference time: {inference_time:.3f} seconds")

    # Handle different output types
    if output_type in ["softmax", "both"]:
        probs = F.softmax(logits, dim=-1)
        top_prob, top_class_id = probs[0].max(dim=-1)
        confidence = f"{top_prob.item():.2%} probability"
    if output_type in ["none", "both"]:
        top_class_id = logits.argmax(-1).item()
        score = logits[0, top_class_id].item()
        confidence = f"{score:.2f} score" if output_type == "none" else f"{confidence} and {score:.2f} score"

    predicted_label = model.config.id2label[top_class_id]
    print(f"Predicted class: {predicted_label} with confidence: {confidence}")

    # Display the image and prediction
    plt.imshow(image)
    plt.title(f"Predicted: {predicted_label} ({confidence})")
    plt.axis('off')
    plt.show()

    # Detailed output
    if output_type == "softmax" or output_type == "both":
        print("Class probabilities:")
        for idx, prob in enumerate(probs[0]):
            class_label = model.config.id2label[idx]
            print(f"{class_label}: {prob.item():.2%}")
    if output_type == "none" or output_type == "both":
        print("Class scores:")
        for idx, score in enumerate(logits[0]):
            class_label = model.config.id2label[idx]
            print(f"{class_label}: {score.item():.2f}")

# Example usage:
classify_image("https://img.freepik.com/premium-psd/half-rotten-apple-isolated-transparent-background-realistic-vector-illustration_220739-12067.jpg?w=1380")
classify_image("https://www.digitaltrends.com/wp-content/uploads/2019/12/tesla-cybertruck-1220.jpg?resize=800%2C418&p=1")
classify_image("https://upload.wikimedia.org/wikipedia/commons/thumb/d/d7/Taylor_Swift_at_the_2023_MTV_Video_Music_Awards_4.png/1200px-Taylor_Swift_at_the_2023_MTV_Video_Music_Awards_4.png")
classify_image("https://assets.bwbx.io/images/users/iqjWHBFdfxIU/iVzchbAHNF1Y/v2/-1x-1.jpg")
classify_image("https://hips.hearstapps.com/popularmechanics/assets/17/20/1494955076-plastic.jpg")
classify_image("https://images.unsplash.com/photo-1507525428034-b723cf961d3e?fm=jpg&q=60&w=3000&ixlib=rb-4.0.3")
classify_image("https://img.jakpost.net/c/2020/08/07/2020_08_07_101893_1596796240._large.jpg")
classify_image("https://live.staticflickr.com/4765/25258187417_684024dcb7.jpg")
```
